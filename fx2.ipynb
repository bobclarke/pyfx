# FXCMPY

## Streaming live prices 

```python
import fxcmpy
import datetime  as dt

from pylab import plt
plt.style.use('seaborn')
%matplotlib inline

con = fxcmpy.fxcmpy(config_file='fxcm.config', server='demo')
```

```python
con.get_instruments_for_candles()
```

### Getting historical prices by specifying the number of results

```python
data = con.get_candles('GBP/JPY', period='D1', number=20)
data
```

### Getting historical prices by specifying the start and end date

```python
start = dt.datetime(2018, 1, 1)
stop = dt.datetime(2018, 1, 30)
data = con.get_candles('EUR/USD', period = 'D1', start = start, stop = stop)
data
#%clear
```

### Subscribing to streaming price updates

```python
con.subscribe_market_data('EUR/USD')
data = con.get_prices('EUR/USD')
con.get_subscribed_symbols()
#con.unsubscribe_market_data('AUD/USD')
```

### plot a chart

```python
data['askclose'].plot(figsize=(10, 6));
```
